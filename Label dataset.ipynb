Use torch.save to save .pt file for small scale

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

image_path = '/content/drive/My Drive/Frechette Lab/CNN_MNP_GUV'
dataset = datasets.ImageFolder(root=image_path, transform=transform)

dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

torch.save(dataset, '/content/drive/My Drive/Frechette Lab/CNN_MNP_GUV/test_dataset.pt')
```
